```python
import pandas as pd
import darts
```

```python
past = pd.read_csv('data/training_values.csv').set_index(['city','year','weekofyear'])
future = pd.read_csv('data/test_values.csv')
labels = pd.read_csv('data/training_labels.csv').set_index(['city','year','weekofyear'])

past = past.join(labels).reset_index()
sj_past = past.loc[past.city == 'sj'].drop(columns ='city')
sj_future = future.loc[future.city == 'sj'].drop(columns ='city')
sj_past.tail(4)
```

```python
display(sj_future.head(2),sj_future.tail(2))
```

```python
past_time_index = pd.date_range(start="1990-04-29", end="2008-04-20",freq='W')
future_time_index = pd.date_range(start="2008-04-27", end="2013-04-23",freq='W')
past_time_index,future_time_index
```

```python
def find_closest_index(datetime_obj,date_time_series):
    time_deltas = (date_time_series-datetime_obj)
    index = time_deltas.days.to_series().abs().argmin()
    return date_time_series[index]

sj_past.week_start_date = pd.to_datetime(sj_past.week_start_date)
sj_future.week_start_date = pd.to_datetime(sj_future.week_start_date)
sj_past['closest_perfekt_interval'] = sj_past.week_start_date.apply(find_closest_index,args=(past_time_index,))
sj_future['closest_perfekt_interval'] = sj_future.week_start_date.apply(find_closest_index,args=(future_time_index,))
```

```python
def interpolate_and_allign_data(data_frame,target_time_index):
    alligned_data = data_frame.groupby('closest_perfekt_interval').mean()
    alligned_data.week_start_date = alligned_data.week_start_date.astype(int).astype(float)
    target = alligned_data.drop(alligned_data.index)
    target = target.reindex(target_time_index.rename('mit_time_index'))
    target = target.astype(alligned_data.dtypes)
    target.update(alligned_data)
    target = target.interpolate()
    return target

sj_past_structured = interpolate_and_allign_data(sj_past,past_time_index)
sj_future_structured = interpolate_and_allign_data(sj_future,future_time_index)
```

```python
sj_future_structured
```

```python
train = darts.TimeSeries.from_dataframe(sj_past_structured,freq='W',)
predict =  darts.TimeSeries.from_dataframe(sj_future_structured,freq='W',)
```

```python
past_and_future_covariates = train.drop_columns(['total_cases']).concatenate(predict)
```

```python
from darts.models import RegressionModel
from sklearn.linear_model import PoissonRegressor

model = RegressionModel(
    #output_chunk_length= 5,
    lags=[-1,-2,-3,-4,-8,-16,-52],
    lags_past_covariates=[-1,-2,-3,-4,-5],
    model=PoissonRegressor()
    )

model.fit([train['total_cases']], past_covariates=[train.drop_columns(['total_cases'])])
```

```python
#benchmarking = model.residuals(train['total_cases'],past_covariates=train.drop_columns(['total_cases']))
```

```python
my_prediction = model.predict(len(predict),past_covariates=past_and_future_covariates)
```
